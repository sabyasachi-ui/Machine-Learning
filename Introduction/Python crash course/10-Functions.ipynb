# Function in Python

### Function is a block of code which executes only when it is called.

### A keyword def is used to initialize the function

```python
def my_addition_func(x,y):     # Defining the function
    print(x+y)
```

```python
my_addition_func(3,5)           # Calling the function so no identation
```

```python
def myadd(x,y,z):
  print(x + y + z)
```

```python
myadd(2,3,4)
```

```python
def mult(a,b,c):
  print(a*b*c)
```

```python
mult(2,5,6)
```

```python
def prime(num):
    for n in range(2, num):              #means for every n in range list( 2 to num)
        if (num % n) == 0:               #our num divided by every n in range list(2 to num)
            print("Not a Prime Number")
            break
    else:
        print("prime Number")

```

```python
prime(19)
```
