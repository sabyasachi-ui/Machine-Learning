```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading dataset from seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head(10)
```

# Distribution Plot

### Setting few parameters for plotting

```python
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
```

```python
plt.figure(figsize = (12,8))
sns.displot(tips['total_bill'], bins = 30)
plt.xlabel('total_bill', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.displot(tips['tip'], bins = 30)
plt.xlabel('tip', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.displot(tips['size'], bins = 30)
plt.xlabel('size', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
tips.head(10)
```

## Count Plot

```python
plt.figure(figsize = (8,5))
sns.countplot(data = tips, x = 'day')
plt.xlabel('day', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
sum(tips['day'] == 'Sun')
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data = tips, x = 'sex')
plt.xlabel('sex', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
sum(tips['sex'] == 'Male')
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data = tips, x = 'smoker')
plt.xlabel('sex', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

# BarPlot

### barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
plt.figure(figsize = (8,5))
sns.barplot(data = tips, x = 'day', y = 'tip')
plt.xlabel('day', fontsize = 15)
plt.ylabel('tip', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.barplot(data = tips, x = 'sex', y = 'tip')
plt.xlabel('day', fontsize = 15)
plt.ylabel('tip', fontsize = 15)
plt.show()
```

# Heatmap

```python
plt.figure(figsize = (6,4))
sns.heatmap(tips[['total_bill', 'tip', 'size']].corr(), cmap = 'coolwarm', annot = True)
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize = (8,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue ='sex', palette = 'Set1')
plt.xlabel('total_bill', fontsize = 15)
plt.ylabel('tip', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue ='day', palette = 'Set1')
plt.xlabel('total_bill', fontsize = 15)
plt.ylabel('tip', fontsize = 15)
plt.show()
```
