# Generating Signals for subplotting

```python
x = np.linspace(0,5,10)
y1 = x ** 2
y2 = x ** 3
```

```python
plt.subplot(2,1,1)  #(rows, columns, position)
plt.plot(x,y1)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.tight_layout()
```

```python
plt.subplot(1,2,1)  #(rows, columns, position)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.tight_layout()
```

# Another Example

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
t = np.linspace(0,10,100)
A = 5
f = 50
sine_wave = A * np.sin(2 * np.pi * f * t)
```

```python
plt.figure(figsize = (16,8))
style.use('dark_background')

plt.subplot(2,1,1)
plt.plot(t,sine_wave)

plt.subplot(2,1,2)
plt.stem(t,sine_wave) # discrete samples of sine wave

plt.show()
```

# Combining Everything together

```python

t = np.arange(0,1,1/300)                     # time 

x1 = 5 * np.sin(2 * np.pi * 2 * t)             # First sinewave

x2 = 2 * np.sin(2 * np.pi * 4 * t)             # Second sinewave

x3 = 7 * np.sin(2 * np.pi * 6 * t)             # Third sinewave

x4 = x1 + x2 + x3                              # Combined sinewave
```

```python
# Plot every sine wave

plt.figure(figsize = (24,14)) # set the size of figure
plt.suptitle('Constructing a Wave with different Sine Waves', fontsize = 40)  # Super title



# Plotting first sine wave at position 01

plt.subplot(2,2,1) 
plt.plot(t,x1,'r',linewidth = 3) 
plt.title("First SineWave", fontsize = 25)
plt.xlabel('time in sec', fontsize = 20)
plt.ylabel('Amplitude', fontsize = 20)

# Plotting second sine wave at position 02

plt.subplot(2,2,2)
plt.plot(t,x2,'y',linewidth = 3)
plt.title("Second SineWave", fontsize = 25)
plt.xlabel('time in sec.', fontsize = 20)
plt.ylabel('Amplitude', fontsize = 20)

# Plotting third sine wave at position 03

plt.subplot(2,2,3)
plt.plot(t,x3,'g',linewidth = 3)
plt.title("Third SineWave", fontsize = 25)
plt.xlabel('time in sec', fontsize = 20)
plt.ylabel('Amplitude', fontsize = 20)

# Plotting fourth wave at position 04

plt.subplot(2,2,4)
plt.plot(t,x4,'m',linewidth = 3)
plt.title("Fourth Wave", fontsize = 25)
plt.xlabel('time in sec', fontsize = 20)
plt.ylabel('Amplitude', fontsize = 20)
plt.show()
```
