```python
import pandas as pd
import numpy as np
```

# Creating Data

```python
X = np.linspace(1,5,10).round(1)
```

```python
X
```

```python
X.shape
```

```python
y = np.arange(0,10)
```

```python
y.shape
```

```python
data = pd.DataFrame({'X': X, 'y' : y})
```

```python
data
```

# Creating New Features using window

```python
win = 2

features = data['X']
label = data['y']
```

```python
features
```

```python
X_Min=[]
X_Max=[]
X_Mean=[]
X_Std=[]

labels=[]

for i in range(len(data) - win):

    X = features[i:i+win]
    y = label[i+win]
    labels.append(y)

    X_Min.append(np.min(X))
    X_Max.append(np.max(X))
    X_Mean.append(np.mean(X))
    X_Std.append(np.std(X))

```

# Create dataframe from features

```python
df_features = pd.DataFrame({'X_Min': X_Min, 'X_Max': X_Max, 'X_Mean': X_Mean, 'X_Std': X_Std,'labels':labels})
```
